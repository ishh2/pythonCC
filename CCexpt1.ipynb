**TEXT HANDLING**

```python
text = "Hello"
print(text.upper())
print(text.lower())
print(text.capitalize())
```

```python
text = "123abc"
print(text.isalpha())   # letters only
print(text.isdigit())   # numbers only
print(text.isalnum())   # letters or numbers
```

```python
text = "Hello, world!"
print(text[0])
print(text[-1])
print(text[0:5])
```

**FILE HANDLING**

```python
#Write to the file
with open('example.txt', 'w') as file:
    file.write("Hello, this is some sample text.\nPython is fun!")

print("Writing done.\n")

#Read from the file
with open('example.txt', 'r') as file:
    contents = file.read()

print("File contents after writing:")
print(contents)
```

```python
import cv2

video = cv2.VideoCapture("cutevideo1.mp4")
frame_count = 0

while True:
    ret, frame = video.read()
    if not ret:
        break
    frame_count += 1

video.release()
print(f"Total frames (manually counted): {frame_count}")
```

```python
video = cv2.VideoCapture("cutevideo1.mp4")

frame_count = 0
while True:
    ret, frame = video.read()
    if not ret:
        break
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(f"frame_{frame_count}.jpg", gray)
    frame_count += 1

video.release()
print(f"{frame_count} frames saved.")
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Open the video file
video = cv2.VideoCapture("cutevideo1.mp4")

# Read and show 10 frames
frame_count = 0
while frame_count < 10:
    ret, frame = video.read()
    if not ret:
        break
    cv2_imshow(frame)
    frame_count += 1

video.release()
```
